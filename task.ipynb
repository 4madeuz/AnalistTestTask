```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import math
import re


platform1_data = pd.read_excel('hotels_parsing_result.xlsx', sheet_name='platform1')
platform2_data = pd.read_excel('hotels_parsing_result.xlsx', sheet_name='platfrom2')
print(len(platform1_data), len(platform2_data))
# Удаление дубликатов
platform1_data.drop_duplicates(inplace=True)
platform2_data.drop_duplicates(inplace=True)
# Преобразование строк в нижний регистр
platform1_data['title'] = platform1_data['title'].str.lower()
platform1_data['city'] = platform1_data['city'].str.lower()
platform2_data['title'] = platform2_data['title'].str.lower()
platform2_data['city'] = platform2_data['city'].str.lower()
# Очистка числовых полей от некорректных значений
numeric_fields = ['rating', 'rating_5', 'star_rating', 'rooms_count', 'lat', 'lon']
platform1_data[numeric_fields] = platform1_data[numeric_fields].apply(pd.to_numeric, errors='coerce')
platform2_data[numeric_fields] = platform2_data[numeric_fields].apply(pd.to_numeric, errors='coerce')
# Удаление строк с отсутствующими значениями в ключевых полях
platform1_data = platform1_data.dropna(subset=['title', 'city', 'lat', 'lon'])
platform2_data = platform2_data.dropna(subset=['title', 'city', 'lat', 'lon'])
#Очистка контактной информации
def clean_contact_info(contact_info):
    if pd.isnull(contact_info):
        return contact_info
    cleaned_contact_info = re.sub('[^0-9a-zA-Z@.]', '', contact_info)
    return cleaned_contact_info

platform1_data['contact_social'] = platform1_data['contact_social'].apply(clean_contact_info)
platform2_data['contact_social'] = platform2_data['contact_social'].apply(clean_contact_info)

print(len(platform1_data), len(platform2_data))
```

```python
def convert_to_list(value):
    if pd.isnull(value):
        return []
    else:
        return [item.strip() for item in value.split(',')]

# Применение функции к полям телефона, электронной почты и веб-сайта
platform1_data['phone'] = platform1_data['phone'].apply(convert_to_list)
platform1_data['email'] = platform1_data['email'].apply(convert_to_list)
platform1_data['website'] = platform1_data['website'].apply(convert_to_list)

platform2_data['phone'] = platform2_data['phone'].apply(convert_to_list)
platform2_data['email'] = platform2_data['email'].apply(convert_to_list)
platform2_data['website'] = platform2_data['website'].apply(convert_to_list)
```

```python
final_platform1 = platform1_data.groupby('title').agg('first').reset_index()
final_platform2 = platform2_data.groupby('title').agg('first').reset_index()
top_10_reviews_platform1 = final_platform1.nlargest(10, 'review_count')
top_10_reviews_platform1
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import numpy as np

# Создание карты
map_center = [platform1_data['lat'].mean(), platform1_data['lon'].mean()]  # Центр карты
mymap = folium.Map(location=map_center, zoom_start=12)

# Создание HeatMap
heat_data = [[row['lat'], row['lon']] for index, row in platform1_data.iterrows()]
HeatMap(heat_data).add_to(mymap)

# Разбивка области карты на квадраты
grid_size_km = 0.01
min_lat, max_lat = platform1_data['lat'].min(), platform1_data['lat'].max()
min_lon, max_lon = platform1_data['lon'].min(), platform1_data['lon'].max()

num_squares_lon = int(np.ceil((max_lon - min_lon) / grid_size_km))

#Индексы квадратов для каждой точки на основе их координат широты и долготы.
lat_indices = ((platform1_data['lat'] - min_lat) / grid_size_km).astype(int)
lon_indices = ((platform1_data['lon'] - min_lon) / grid_size_km).astype(int)

# Отбираем уникальные индексы квадратов, в которых есть хотя бы одна точка
unique_square_indices, counts = np.unique(lat_indices * num_squares_lon + lon_indices, return_counts=True)


# Выбор квадрата с максимальным количеством точек
max_square_index = unique_square_indices[np.argmax(counts)]
lat_idx = max_square_index // num_squares_lon
lon_idx = max_square_index % num_squares_lon

max_square_lat = min_lat + lat_idx * grid_size_km
max_square_lon = min_lon + lon_idx * grid_size_km

# Отображение всех квадратов на карте
for unique_index in unique_square_indices:
    lat_idx = unique_index // num_squares_lon
    lon_idx = unique_index % num_squares_lon

    lat = min_lat + lat_idx * grid_size_km
    lon = min_lon + lon_idx * grid_size_km

    # Отображение квадрата на карте
    square_polygon = folium.Polygon(
        [(lat, lon),
         (lat + grid_size_km, lon),
         (lat + grid_size_km, lon + grid_size_km),
         (lat, lon + grid_size_km)],
        color='gray', fill=True, fill_color='gray', fill_opacity=0.1
    )
    square_polygon.add_to(mymap)

# Выделение выбранного квадрата цветом
selected_square_polygon = folium.Polygon(
    [(max_square_lat, max_square_lon),
     (max_square_lat + grid_size_km, max_square_lon),
     (max_square_lat + grid_size_km, max_square_lon + grid_size_km),
     (max_square_lat, max_square_lon + grid_size_km)],
    color='blue', fill=True, fill_color='blue', fill_opacity=0.2
)
selected_square_polygon.add_to(mymap)

# Отображение карты
mymap.save('heatmap_with_max_square.html')  # Сохранение карты в HTML-файл
mymap
```
